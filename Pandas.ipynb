```python
#import pandas
import pandas as pd

#import the Pokemon info
df = pd.read_csv('pokemon_data.csv')
print(df.head())
```

```python
#describing data
df.describe()

#sorting
df.sort_values('Name', ascending=False)
df.sort_values(['HP', 'Name'], ascending=[True, False])
```

```python
#making column 'Total' and setting it to horizontal sum
df['Total'] = df.iloc[:, 4:10].sum(axis=1)

#moving total to the left of HP
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
```

```python
#export to csv
df.to_csv('modified.csv', index=False)
```

```python
#filtering data
new_df = df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison') & (df['HP'] > 70)]

#reset the index, drop old indexes
new_df.reset_index(drop=True)

#names that do not contain Mega
df.loc[~df['Name'].str.contains('Mega')]

#using regular expressions and flags
#re.I flag is ignore case
#import flags
import re
df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)]

#name starts with pi
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

```python
#conditional changes; change Fire to Flamer
df.loc[df['Type 1'] == 'Fire', 'Type 1'] = 'Flamer'

#make new column Legendary and set Total == 600 to true 
df.loc[df['Total'] >= 600, 'Legendary'] = True
df
```

```python
#using groupby to find aggregate statistics
#mean
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)

#sum
df.groupby(['Type 1']).sum()

#count
df.groupby(['Type 1', 'Type 2']).count()
```
