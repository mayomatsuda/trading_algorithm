```python
#importing numpy
import numpy as np

#assign 'Adj Close' to 'daily_close'
daily_close = msft_data[['Adj_Close']]

#returns as fractional change
daily_return = daily_close.pct_change()

#replacing NA values with 0
daily_return.fillna(0, inplace=True)

print(daily_return)
```

```python
#resample and get last day of trading
mdata = msft_data.resample('M').apply(lambda x: x[-1])[['Adj_Close']]

#return as fractional change
monthly_return = mdata.pct_change()

#replace NA values with 0
monthly_return.fillna(0, inplace=True)

print(monthly_return)
```

```python
#assigning adjusted closing prices
adj_price = msft_data['Adj_Close']

#calculate the moving average
mav = adj_price.rolling(window=50).mean()

print(mav[-10:])
```

```python
#import matplotlib package to see the plot
import matplotlib.pyplot as plt

#plot adjusted closing prices
adj_price.plot()
```

```python
#plot rolling average
mav.plot()
```

```python
#momentum-based long-only Simple Moving Average Crossover (SMAC) strategy
#momentum is the total return of stock over last n months (lookback period)
#3 types of lookback periods: short term, intermediate term, long term
#buy signal when shorter lookback rolling mean overshoots longer one
#sell signal when longer lookback rolling mean overshoots shorter one

#initialize short and long lookback periods
short_lb = 50
long_lb = 120

#initialize a new DataFrame called signal_df with a signal column
signal_df = pd.DataFrame(index=msft_data.index)
signal_df['signal'] = 0.0

#create a short simple moving average over short lookback period
signal_df['short_mav'] = msft_data['Adj_Close'].rolling(window=short_lb, min_periods=1, center=False).mean()

#create a long simple moving average over long lookback period
signal_df['long_mav'] = msft_data['Adj_Close'].rolling(window=long_lb, min_periods=1, center=False).mean()

#generate the signals based on the conditional statement
signal_df['signal'][short_lb:] = np.where(signal_df['short_mav'][short_lb:] > signal_df['long_mav'][short_lb:], 1.0, 0.0)

#create the trading orders based on the positions column
signal_df['positions'] = signal_df['signal'].diff()
signal_df[signal_df['positions'] == -1.0]
```

```python
#initialize a plot
fig = plt.figure()

#add a subplot and label for y-axis
plt1 = fig.add_subplot(111, ylabel='Price in $')

#plot adjusted close
msft_data['Adj_Close'].plot(ax=plt1, color='r', lw=2)

#plot short and long lookback moving averages
signal_df[['short_mav', 'long_mav']].plot(ax=plt1, lw=2, figsize=(12,8))

#plot sell signals
plt1.plot(signal_df.loc[signal_df.positions == -1.0].index, signal_df.short_mav[signal_df.positions == -1.0], 'v', markersize=10, color='k')

#plot buy signals
plt1.plot(signal_df.loc[signal_df.positions == 1.0].index, signal_df.short_mav[signal_df.positions == 1.0], '^', markersize=10, color='m')

plt.show()
```
